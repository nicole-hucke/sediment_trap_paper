```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns

samples = pd.read_csv('../../data/discharge/rating_curve_samples.csv')
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(6,6))
sns.scatterplot(data=samples, y='Q (cfs)', x='Stage (cm)', s=90,
                color='olivedrab', edgecolor='olivedrab', alpha=0.7)
# create fit line
stage_range = np.linspace(samples['Stage (cm)'].min(), samples['Stage (cm)'].max(), 100)
#stage_range_linear = np.linspace(19.682911, samples['Stage (cm)'].max(), 100)
q_fit = 0.000000006 * (stage_range) ** 6.1991
#q_fit_reg = 0.5757 * stage_range_linear - 10.7
# add fitted line
plt.plot(stage_range, q_fit, '--', color='olivedrab', linewidth=2)
#plt.plot(stage_range_linear, q_fit_reg, '--', color='olivedrab', linewidth=2)
# make the grid lines a bit more subtle
plt.grid(color='gray', linestyle='-', linewidth=0.5, alpha=0.6)
plt.xlabel('Stage (cm)', fontsize=14)
plt.ylabel('Discharge (cfs)', fontsize=14)
plt.xscale('log')
plt.yscale('log')
plt.savefig('rating_curve_log.png', dpi=500)
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(6,6))
sns.scatterplot(data=samples, y='Q (cfs)', x='Stage (cm)', s=90,
                color='olivedrab', edgecolor='olivedrab', alpha=0.7)
# create fit line
stage_range = np.linspace(samples['Stage (cm)'].min(), 19.682911, 100)
stage_range_linear = np.linspace(19.682911, samples['Stage (cm)'].max(), 100)
q_fit = 0.000000006 * (stage_range) ** 6.1991
q_fit_reg = 0.5757 * stage_range_linear - 10.7
# add fitted line
plt.plot(stage_range, q_fit, '--', color='olivedrab', linewidth=2)
plt.plot(stage_range_linear, q_fit_reg, '--', color='olivedrab', linewidth=2)
plt.grid(color='gray', linestyle='-', linewidth=0.5, alpha=0.6)
plt.xlabel('Stage (cm)', fontsize=14)
plt.ylabel('Discharge (cfs)', fontsize=14)

plt.savefig('rating_curve_piecewise.png', dpi=500)
plt.show()
```
