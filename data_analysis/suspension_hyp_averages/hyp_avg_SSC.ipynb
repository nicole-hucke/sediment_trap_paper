### Calculating Turbidity Threshold

#### Percentile threshold 

Functions

```python
def merge_Q_turb(df_Q: pd.DataFrame, df_T: pd.DataFrame, q_col_name: str, turb_col_name: str, how: str = "inner",):
    # checking if the columns exist in the dataframes
    if q_col_name not in df_Q.columns:
        raise KeyError(f"'{q_col_name}' not found in df_Q columns")
    if turb_col_name not in df_T.columns:
        raise KeyError(f"'{turb_col_name}' not found in df_T columns")
    # merging the dataframes
    merged = (df_Q[[q_col_name]].merge(df_T[[turb_col_name]], left_index=True, right_index=True, how=how))
    return merged

def q_cutoff_from_baseflow(merged_df: pd.DataFrame, q_col: str, baseflow_quantile: float = 0.50,): 
    # filtering the merged dataframe to get the baseflow values
    return merged_df[q_col].quantile(baseflow_quantile)

def turbidity_threshold(merged_df: pd.DataFrame, Q_cutoff: float, q_col: str, turb_col: str, percentile: float = 99.0,):
    # filtering the turbidity values based on the Q cutoff
    base_turb = merged_df.loc[merged_df[q_col] <= Q_cutoff, turb_col].dropna() 
    # calculating the percentile of the filtered turbidity values
    return np.percentile(base_turb, percentile)
```

### Getting Exceedance Periods

```python
def get_exceedance_periods(df, turb_col, threshold, season_label):
    exceedances = df[turb_col] > threshold
    periods = []
    in_event = False
    for i in range(len(exceedances)):
        if exceedances.iloc[i] and not in_event:
            start = df.index[i]
            in_event = True
        elif not exceedances.iloc[i] and in_event:
            end = df.index[i]
            periods.append((season_label, start, end))
            in_event = False
    if in_event:
        periods.append((season_label, start, df.index[-1]))
    return periods
```

```python
# get periods for spring and summer
spring_threshold = 14.13
summer_threshold = 5.55

spring_periods = get_exceedance_periods(sp_SSC, 'Turbidity (NTU)', spring_threshold, 'spring')
summer_periods = get_exceedance_periods(sm_SSC, 'Turbidity (FNU)', summer_threshold, 'summer')
# combine and export
all_periods = spring_periods + summer_periods
df_periods = pd.DataFrame(all_periods, columns=['season', 'start', 'end'])
percentile_periods = df_periods.copy()

# print the result
print(df_periods)
# export to CSV
df_periods.to_csv('exceedance_periods/turb_exceedance_percentile.csv', index=False)

# calculate total exceedance time
df_periods['start'] = pd.to_datetime(df_periods['start'])
df_periods['end'] = pd.to_datetime(df_periods['end'])
# calculate duration for each row
df_periods['duration'] = df_periods['end'] - df_periods['start']
# group by season and sum durations
duration_by_season = df_periods.groupby('season')['duration'].sum()
# convert to minutes
duration_minutes = duration_by_season.apply(lambda x: x.total_seconds() / 60)
# print result
for season, minutes in duration_minutes.items():
    print(f"Total exceedance time in {season}: {minutes:.2f} minutes")
```

### Getting HF time series
Using the sample threshold, as this is what I used as final metric in my paper

```python
spring_HF = pd.read_csv('hyporheic_fluxes/spring_HF.csv', index_col=0, parse_dates=True)
summer_HF = pd.read_csv('hyporheic_fluxes/summer_HF.csv', index_col=0, parse_dates=True)
# interpolate to get 1m intervals
spring_HF = spring_HF.resample('1min').interpolate()
summer_HF = summer_HF.resample('1min').interpolate()

# build HF time series for each exceedance period and plot
df_periods[['start', 'end']] = df_periods[['start', 'end']].apply(pd.to_datetime)

hf_sources = {'spring': spring_HF, 'summer': summer_HF}
period_slices = []

for idx, row in df_periods.iterrows():
    season = str(row['season']).lower()
    hf_df = hf_sources.get(season)
    if hf_df is None:
        continue  # skip unknown seasons

    window = hf_df.loc[row['start']:row['end']].copy()
    if window.empty:
        continue

    window['season'] = season
    window['period_id'] = f'{season}_{idx}'
    window['period_start'] = row['start']
    window['period_end'] = row['end']
    period_slices.append(window)

hf_period_ts = pd.concat(period_slices).sort_index()
display(hf_period_ts.head())

fig, ax = plt.subplots(figsize=(12, 5))
hf_period_ts.filter(regex=r'^T\d+$').plot(ax=ax, alpha=0.8)
for _, r in df_periods.iterrows():
    ax.axvspan(r['start'], r['end'], color='gray', alpha=0.15)
ax.set_title('Hyporheic Flux During Exceedance Periods')
ax.set_ylabel('HF')
ax.set_xlabel('Date')   
ax.legend(loc='upper right')
plt.show()

# export as CSV
hf_period_ts.to_csv('hyporheic_fluxes/hf_during_exceedance_periods.csv')
```

SSC Rating Curve Threshold

```python
# filter to just spring exceedance periods
spring_periods = rating_curve_periods[rating_curve_periods['season'] == 'spring']
summer_periods = rating_curve_periods[rating_curve_periods['season'] == 'summer']

# create a list to collect all rows that fall within the exceedance periods
rc_sp_exceedance = pd.DataFrame()
rc_sm_exceedance = pd.DataFrame()

# loop through each exceedance period and get the data
# spring 
for _, row in spring_periods.iterrows():
    start = row['start']
    end = row['end']
    # filter the data to get the rows that fall within the exceedance period
    period_data = spring_data[(spring_data.index >= start) & (spring_data.index <= end)]
    # append the data to the list
    rc_sp_exceedance = pd.concat([rc_sp_exceedance, period_data])

# summer
for _, row in summer_periods.iterrows():
    start = row['start']
    end = row['end']
    # filter the data to get the rows that fall within the exceedance period
    period_data = summer_data[(summer_data.index >= start) & (summer_data.index <= end)]
    # append the data to the list
    rc_sm_exceedance = pd.concat([rc_sm_exceedance, period_data])

# calculate the weighted mean for each probe
weighted_means = {}

# spring
for col in ['T1', 'T2', 'T3', 'T5', 'T6', 'T7', 'T8']:
    hf_values = rc_sp_exceedance[col]
    turbidity_values = rc_sp_exceedance['Turbidity (NTU)']
    weighted_avg = (hf_values * turbidity_values).sum() / turbidity_values.sum()
    weighted_means[col] = weighted_avg
weighted_means_df = pd.DataFrame.from_dict(weighted_means, orient='index', columns=['Weighted HF (Spring)'])
print(weighted_means_df)
weighted_means_df.to_csv('weighted_avg_HF/ratingcurve_weighted_spring_HF.csv', index=True)

# summer
for col in ['T1', 'T2', 'T3', 'T5', 'T6', 'T7', 'T8']:
    hf_values = rc_sm_exceedance[col]
    turbidity_values = rc_sm_exceedance['Turbidity (FNU)']
    weighted_avg = (hf_values * turbidity_values).sum() / turbidity_values.sum()
    weighted_means[col] = weighted_avg
weighted_means_df = pd.DataFrame.from_dict(weighted_means, orient='index', columns=['Weighted HF (Summer)'])
print(weighted_means_df)
weighted_means_df.to_csv('weighted_avg_HF/ratingcurve_weighted_summer_HF.csv', index=True)
```
