### Wilcoxon signed-rank test on standard deviation differences (Δσ)

```python
import numpy as np
from scipy.stats import wilcoxon, ttest_rel

# standard deviations (including T2 and T3)
sigma_spring = np.array([0.05763, 0.04136, 0.05916, 0.02016, 0.01348, 0.09615, 0.01658])
sigma_summer = np.array([0.27808, 0.03804, 0.03680, 0.06306, 0.03882, 0.05678, 0.02984])
```

Wilcoxon rank-sum test (paired) - With T2 and T3 

```python
stat, p_value = wilcoxon(
    sigma_summer,
    sigma_spring,
    alternative="greater"  # tests whether summer > spring
)

print(f"Wilcoxon statistic: {stat:.3f}")
print(f"P-value: {p_value:.4f}")
```

Wilcoxon rank-sum test (paired) - Without T2 and T3 

```python
# standard deviations (including T2 and T3)
sigma_spring = np.array([0.05763, 0.02016, 0.01348, 0.09615, 0.01658])
sigma_summer = np.array([0.27808, 0.06306, 0.03882, 0.05678, 0.02984])
```

```python
stat, p_value = wilcoxon(
    sigma_summer,
    sigma_spring,
    alternative="greater"  # tests whether summer > spring
)

print(f"Wilcoxon statistic: {stat:.3f}")
print(f"P-value: {p_value:.4f}")
```

T-Test

```python
stat, p_value = ttest_rel(
    sigma_summer,
    sigma_spring) 

print(f"t-test statistic: {stat:.3f}")
print(f"P-value: {p_value:.4f}")
```
